```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_csv("/content/Country.csv", index_col ="name" )
df1.drop(["Belgium", "Switzerland","Scotland","Portugal","Poland","Netherlands"], axis = 0, inplace = True)
df1
  
```

```python
df2=pd.read_csv("/content/League.csv")
df2
```

```python
df2.rename(columns = {'name':'League Name'}, inplace = True)
print(df2)
```

```python
df3=pd.merge(left = df1, right =df2 , how = 'inner')
df3
```

```python
df4=pd.read_csv("/content/Team.csv")
df4
```

```python
df5=pd.merge(left = df3, right =df4 , how = 'right')
df5
```

```python
df6=pd.read_csv("/content/player.csv.xls")
df6
```

```python
df8=pd.merge(left = df4, right =df6 , how = 'right')
df8
```

```python
df7=pd.read_csv("/content/Match.csv")
df7
```

```python
df9=pd.merge(left = df6, right =df7 , how = 'right')
df9
```

```python
df9=pd.read_csv("/content/team_attributes.csv.xls")
df9
```

```python
df10=pd.merge(left = df7, right =df9 , how = 'outer')
df10
```

```python
M1=pd.merge(left=df5,right=df6,on='country_id',how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
M1
```

```python
M2=pd.merge(left=df,right=df1,left_on='id',right_on="id",how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
M2
```

```python
M3=pd.merge(left=df2,right=df3,left_on='id',right_on="id",how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
M3
```

```python
M4=pd.merge(left=M2,right=M3,left_on='id',right_on="id",how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
M4
```

```python
M5=pd.merge(left=M4,right=M1,left_on='id',right_on="id",how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
```

```python
M6=pd.merge(left=df4,right=M5,left_on='id',right_on="id",how='right',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
M6
```

```python
M6=M6[['name','team_long_name','team_short_name','season','leaguename','home_team_goal','away_team_goal','buildUpPlaySpeedClass','buildUpPlayPassingClass']]
```

```python
import matplotlib.pyplot as plt
sns.barplot(x='name',y='home_team_goal',data=M6)
plt.show
plt.xlabel('Countryname')
plt.ylabel('Number of goals')
```

```python
import matplotlib.pyplot as plt
sns.barplot(x='team_long_name',y='home_team_goal',data=M6)
plt.show
plt.xlabel('Team')
plt.ylabel('Number of goals')
```

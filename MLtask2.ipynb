```python
import pandas as pd
```

```python
df=pd.read_csv("/content/data.csv")
df
```

```python
#Find songs with value for energy column greater than 0.4 and display values for song_title column.
df1=df[df["energy"] > 0.4]
df4=df1['song_title']
df4
```

```python
#For songs with duration duration_ms value less than 300000 ms (300 s), find Top 10 “fastest” 
#songs( songs with highest values for tempo column ) and display values for song_title, artistcolumns.
df2 = df[df['duration_ms'] < 300000]
df3 = df2.sort_values('tempo',ascending=False)[['song_title', 'artist']].head(10)
df3
```

```python
#Find songs with value for artist column = Drake and value for energy column greater than 0.5 and display values for song_title column.
df5=df[df.artist=='Drake']
df6=df5[df.energy>0.5]
df7=df6['song_title']
df7
```

```python
#Create a dataframe of the songs that rank among top 10 for the instrumentalness metric. This dataframe should only contain the instrumentalness, song title and artist columns.
df8=df.sort_values('instrumentalness',ascending=False)[['song_title', 'artist','instrumentalness']].head(10)
df8
```
